## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 2.2 | </b></span><span style="color:#1d84b5"><b>Visualize Data</b></span>


```python
for class_name in classes:
    # Create the full path to the class directory
    class_path = os.path.join(base_dir, class_name)

    # List all images in the current class directory
    images = os.listdir(class_path)

    # Randomly select 5 images from the class directory
    random_images = random.choices(images, k=5)

    # Set up a subplot with 1 row and 5 columns for displaying images
    fig, ax = plt.subplots(1, 5, figsize=(10, 1.5))

    for i in range(5):
        # Set the title for the current class
        plt.suptitle(class_name, fontsize=15, fontweight='bold')

        # Create the full path to the selected image
        img_path = os.path.join(class_path, random_images[i])

        # Load the image using PyTorch
        img = read_image(img_path)  # Load image as a tensor
        img = transforms.ToPILImage()(img)  # Convert tensor to PIL image for displaying

        # Display the image in the subplot
        ax[i].imshow(img)
        ax[i].axis('off')  # Turn off axis

    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()  # Show the plot
```

```python
# Count images per class and create a bar plot
# Count images per class
counts = [len(os.listdir(os.path.join(base_dir, class_name))) for class_name in classes] 

# Create a figure for the bar plot
plt.figure(figsize=(12, 5))

# Create a vertical bar plot with classes on the X-axis and counts on the Y-axis
ax = sns.barplot(x=classes, y=counts)

# Annotate each bar with the corresponding count
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10)

# Set the title of the plot
plt.title('Number of images per class', fontsize=25, fontweight='bold')

# Set the X-axis label
plt.xlabel('Classes', fontsize=15)

# Set the Y-axis label
plt.ylabel('Number of samples', fontsize=15)

# Rotate the X-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Batch size for data loaders
batch_size = 32

# Function to create data loaders for different datasets
def create_data_loader(dataset_path, transform, batch_size, dataset_type):
    # Read images as a dataset
    dataset = torchvision.datasets.ImageFolder(dataset_path, transform=transform)
    
    # Create a Data Loader
    data_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
    
    # Print dataset information
    print(colored(f'{dataset_type} Folder :\n', 'green', attrs=['bold']))
    print(dataset)
    
    return data_loader

# Transform for data augmentation and normalization (replace 'transform' with desired transformations)
transform = transforms.Compose([
    # Resize images to 224x224 for consistency
    transforms.Resize((224, 224)),  
    # Convert image to tensor
    transforms.ToTensor(),  
    # Normalize using ImageNet mean and std
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  
])

# Create Data Loaders for train, validation, and test datasets
train_loader = create_data_loader(os.path.join(Out_dir, 'train'), transform, batch_size, 'Train')
val_loader = create_data_loader(os.path.join(Out_dir, 'val'), transform, batch_size, 'Validation')
test_loader = create_data_loader(os.path.join(Out_dir, 'test'), transform, batch_size, 'Test')
```

```python
# Setup Data Parallelism for Multi-GPU Training
# DataParallel is used to wrap the model for utilizing multiple GPUs.
# PyTorch's DataParallel splits the input data across the GPUs, processes it,
# and then combines the results to a single output.

# Define the CNN model class
class CNN(nn.Module):
    def __init__(self, num_classes):
        # Call the parent class (nn.Module) initializer
        super(CNN, self).__init__()
        
        # Define the convolutional layers
        self.conv_layers = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3),  # First convolutional layer
            nn.ReLU(),  # ReLU activation
            nn.MaxPool2d(kernel_size=(2, 2)),  # Max pooling
            nn.BatchNorm2d(32),  # Batch normalization
            
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3),  # Second convolutional layer
            nn.ReLU(),  # ReLU activation
            nn.MaxPool2d(kernel_size=(2, 2)),  # Max pooling
            nn.BatchNorm2d(64),  # Batch normalization
            
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3),  # Third convolutional layer
            nn.ReLU(),  # ReLU activation
            nn.MaxPool2d(kernel_size=(2, 2)),  # Max pooling
            nn.BatchNorm2d(128),  # Batch normalization
        )
        
        # Define the fully connected (dense) layers
        self.dense_layers = nn.Sequential(
            nn.Linear(128 * 26 * 26, 128),  # First fully connected layer
            nn.ReLU(),  # ReLU activation
            nn.Linear(128, 64),  # Second fully connected layer
            nn.ReLU(),  # ReLU activation
            nn.Linear(64, num_classes)  # Final output layer
        )

    def forward(self, X):
        # Pass the input through the convolutional layers
        out = self.conv_layers(X)
        
        # Flatten the output from the convolutional layers to pass into the dense layers
        out = out.view(out.size(0), -1)  # Flatten the output tensor
        
        # Pass the flattened output through the dense layers
        out = self.dense_layers(out)
        return out

# Correct the missing variable by defining 'num_classes'
num_classes = 5  # Number of output classes in the model

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  # Select device, use GPU if available

# Instantiate and wrap the model in DataParallel to utilize multiple GPUs
model = CNN(num_classes)
if torch.cuda.device_count() > 1:
    print("Using multiple GPUs for training.")
    model = nn.DataParallel(model)  # Wrap the model to use multiple GPUs
model.to(device)  # Move the model to the selected device (GPU or CPU)
```

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 4.2 | </b></span><span style="color:#1d84b5"><b>Train Model</b></span>


```python
# Create Loss function and Optimizer
# Define the cross-entropy loss for classification tasks
criterion = nn.CrossEntropyLoss()

# Define the Adam optimizer with a learning rate of 0.001
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Define number of epochs
num_epochs = 10

# Variables to store loss and accuracy for plotting
# Store training and validation loss and accuracy for each epoch
train_losses = np.zeros(num_epochs)
val_losses = np.zeros(num_epochs)
train_accs = np.zeros(num_epochs)
val_accs = np.zeros(num_epochs)

# Training loop
for it in range(num_epochs):
    # Set model to training mode
    model.train()
    
    # Start time for epoch
    t0 = datetime.now()
    
    # List to store batch-wise training and validation losses
    train_loss = []
    val_loss = []
    
    # Counters for correct training and validation predictions
    n_correct_train = 0
    n_total_train = 0

    # Training phase
    for images, labels in train_loader:
        # Move images and labels to the selected device
        images = images.to(device)
        labels = labels.to(device)

        # Zero the gradients from the previous iteration
        optimizer.zero_grad()

        # Forward pass to get predictions
        y_pred = model(images)

        # Calculate loss
        loss = criterion(y_pred, labels)

        # Backward pass to calculate gradients
        loss.backward()

        # Update model parameters
        optimizer.step()

        # Append current batch loss to list
        train_loss.append(loss.item())

        # Get class index with highest probability
        _, prediction = torch.max(y_pred, 1)

        # Increment correct predictions count
        n_correct_train += (prediction == labels).sum().item()
        
        # Increment total number of samples
        n_total_train += labels.shape[0]

    # Average training loss for the epoch
    train_losses[it] = np.mean(train_loss)

    # Training accuracy for the epoch
    train_accs[it] = n_correct_train / n_total_train

    # Validation phase
    # Set model to evaluation mode
    model.eval()
    
    # Counters for correct validation predictions
    n_correct_val = 0
    n_total_val = 0
    
    # Disable gradient computation during validation
    with torch.no_grad():
        for images, labels in val_loader:
            # Move images and labels to the selected device
            images = images.to(device)
            labels = labels.to(device)

            # Forward pass to get predictions
            y_pred = model(images)

            # Calculate validation loss
            loss = criterion(y_pred, labels)

            # Append current batch loss to list
            val_loss.append(loss.item())

            # Get class index with highest probability
            _, prediction = torch.max(y_pred, 1)

            # Increment correct predictions count
            n_correct_val += (prediction == labels).sum().item()
            
            # Increment total number of samples
            n_total_val += labels.shape[0]

    # Average validation loss for the epoch
    val_losses[it] = np.mean(val_loss)

    # Validation accuracy for the epoch
    val_accs[it] = n_correct_val / n_total_val

    # Calculate duration of epoch
    dt = datetime.now() - t0

    # Print results for the current epoch
    print(f'Epoch [{it + 1}/{num_epochs}] -> Train Loss: {train_losses[it]:.4f},Train Acc: {train_accs[it]:.4f} | Val Loss: {val_losses[it]:.4f},Val Acc: {val_accs[it]:.4f} | Duration: {dt}')
```
