## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 2.2 | </b></span><span style="color:#1d84b5"><b>Visualize Data</b></span>


```python
for class_name in classes:

    # Create the full path to the class directory

    class_path = os.path.join(base_dir, class_name)



    # List all images in the current class directory

    images = os.listdir(class_path)



    # Randomly select 5 images from the class directory

    random_images = random.choices(images, k=5)



    # Set up a subplot with 1 row and 5 columns for displaying images

    fig, ax = plt.subplots(1, 5, figsize=(10, 1.5))

    

    for i in range(5):

        # Set the title for the current class

        plt.suptitle(class_name, fontsize=15, fontweight='bold')

        

        # Create the full path to the selected image

        img = os.path.join(class_path, random_images[i])

        

        # Load the image using Keras

        img = keras.utils.load_img(img)

        

        # Display the image in the subplot

        ax[i].imshow(img)

        ax[i].axis('off')  # Turn off axis



    plt.tight_layout()  # Adjust layout to prevent overlap
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `classes` is a list of class names and `base_dir` is defined
counts = [len(os.listdir(os.path.join(base_dir, class_name))) for class_name in classes]  # Count images per class

# Create a figure for the bar plot
plt.figure(figsize=(12, 5))

# Create a vertical bar plot with classes on the X-axis and counts on the Y-axis
ax = sns.barplot(x=classes, y=counts)

# Annotate each bar with the corresponding count
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10)

# Set the title of the plot
plt.title('Number of images per class', fontsize=25, fontweight='bold')

# Set the X-axis label
plt.xlabel('Classes', fontsize=15)

# Set the Y-axis label
plt.ylabel('Number of samples', fontsize=15)

# Rotate the X-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping

# Show the plot

plt.show()
```

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 3.1 | </b></span><span style="color:#1d84b5"><b>Create DataFrame</b></span>


<div style="background-color:#0A2342; padding: 20px 10px 10px 10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.1); border:0px solid #0A2342; text-align:left; color: #f2f2f2;">
    <p style="font-size:15px; font-family:tahoma; line-height: 2em; text-indent: 20px;">➡️ Here, create a DataFrame with columns: <b>File_path, Label</b> stored in each row.</p>
    <p style="font-size:15px; font-family:tahoma; line-height: 2em; text-indent: 20px;">➡️ And shuffle it.</p>

</div>


```python
# Create a list of tuples that contain image paths and their class labels

my_list = []  # Initialize an empty list to store tuples of file paths and class labels

for class_name in classes:

    # Create the full path to the class directory

    class_path = os.path.join(base_dir, class_name)

    

    # List all files in the class directory

    files = os.listdir(class_path)

    

    for file in files:

        # Create the full file path

        file_path = os.path.join(class_path, file)

        

        # Append a tuple of (file_path, class_name) to the list

        my_list.append((file_path, class_name))



# Convert the list to a pandas DataFrame with specified column names

df = pd.DataFrame(my_list, columns=['file_path', 'label'])



# Shuffle the dataframe rows randomly

df = df.sample(frac=1).reset_index(drop=True)



# Display the first few rows of the dataframe

df.head()
```

```python
# Split the DataFrame into train, validation, and test DataFrames

train_df = df[:60000]         # First 60,000 samples for training

valid_df = df[60000:67500]    # Next 7,500 samples for validation

test_df = df[67500:]          # Remaining samples for testing



# Reset indexes of DataFrames

train_df.reset_index(inplace=True, drop=True)  # Reset index for training DataFrame

valid_df.reset_index(inplace=True, drop=True)  # Reset index for validation DataFrame

test_df.reset_index(inplace=True, drop=True)    # Reset index for testing DataFrame



# Print the number of samples in each DataFrame with colored output

print(colored(f'Number of samples in train DataFrame : {len(train_df)}', 'red', attrs=['bold']))

print(colored(f'Number of samples in validation DataFrame : {len(valid_df)}', 'red', attrs=['bold']))

print(colored(f'Number of samples in test DataFrame : {len(test_df)}', 'red', attrs=['bold']))
```

<div style="background-color:#0A2342; padding: 20px 10px 10px 10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.1); border:0px solid #0A2342; text-align:left; color: #f2f2f2;">
    <p style="font-size:15px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        ➡️ In this part, I use a <b>Generator</b> to fetch batches of images from it for the model.
    </p>
    <p style="font-size:15px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        ➡️ Additionally, I apply slight data augmentation in this generator to expand the dataset.
    </p>

</div>


```python
# Set the image size and batch size

image_size = (224, 224)  # Define the target image size for the model

batch_size = 16           # Define the number of images to be processed in one batch



# Create an ImageDataGenerator object with rescale options for image preprocessing

datagen = ImageDataGenerator(

    rescale=1./255  # Rescale pixel values to the range [0, 1]

)



# Create a generator for the training data

train_generator = datagen.flow_from_dataframe(

    train_df,            # DataFrame containing training data

    x_col='file_path',   # Column name for image file paths

    y_col='label',       # Column name for labels

    target_size=image_size,  # Resize images to the target size

    batch_size=batch_size,    # Number of images per batch

    class_mode='categorical', # Specify that this is a multi-class classification problem

    shuffle=True          # Shuffle training data for better randomness

)



# Create a generator for the validation data

valid_generator = datagen.flow_from_dataframe(

    valid_df,           # DataFrame containing validation data

    x_col='file_path',  # Column name for image file paths

    y_col='label',      # Column name for labels

    target_size=image_size,  # Resize images to the target size

    batch_size=batch_size,    # Number of images per batch

    class_mode='categorical', # Specify that this is a multi-class classification problem

    shuffle=False        # Do not shuffle validation data

)



# Create a generator for the test data

test_generator = datagen.flow_from_dataframe(

    test_df,           # DataFrame containing test data

    x_col='file_path',  # Column name for image file paths

    y_col='label',      # Column name for labels

    target_size=image_size,  # Resize images to the target size

    batch_size=batch_size,    # Number of images per batch

    class_mode='categorical', # Specify that this is a multi-class classification problem

    shuffle=False        # Do not shuffle test data

)
```

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 4.1 | </b></span><span style="color:#1d84b5"><b>Build Model</b></span>


```python
# 1. Configure Memory Growth for GPUs
# Retrieve the list of available GPUs. 
# By default, TensorFlow attempts to allocate all GPU memory at the start, which can lead to memory issues.
# Setting memory growth allows TensorFlow to allocate only as much GPU memory as needed at a time,
# and grow the memory usage as necessary.
gpus = tf.config.list_physical_devices('GPU')
if gpus:  # Check if there are any GPUs available
    try:
        # Loop through each GPU and set memory growth to True.
        # This approach prevents TensorFlow from consuming all GPU memory initially.
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("Memory growth is set for GPUs.")
    except RuntimeError as e:
        # Memory growth must be set before GPUs have been initialized.
        # If GPUs are already in use, a RuntimeError will occur.
        print(e)

# 2. Setup Multi-GPU Training Strategy with MirroredStrategy
# MirroredStrategy is used for synchronous training across multiple GPUs,
# where each GPU receives the same copy of the model and data. Gradients from each GPU 
# are combined (synchronized) and averaged after each training step.
# This approach ensures that model training is balanced and synchronized across all GPUs.
strategy = tf.distribute.MirroredStrategy()  # Initialize the strategy for distributing the model
```

```python
# Set the input shape for the model

input_shape = (224, 224, 3)  # Input shape for the model: height, width, number of color channels (RGB)

# 3: Build and compile the model within the strategy scope
# Using the strategy.scope() ensures that the model is distributed across multiple GPUs
with strategy.scope():
    model = Sequential()  # Initialize a Sequential model
    
    # Add the first convolutional layer
    
    model.add(Conv2D(32, 3, activation='relu', input_shape=input_shape))  # 32 filters, 3x3 kernel size, ReLU activation
    
    model.add(MaxPooling2D(2, 2))  # Max pooling layer to downsample the feature maps
    
    
    
    model.add(keras.layers.BatchNormalization())  # Normalize the activations of the previous layer
    
    
    
    # Add the second convolutional layer
    
    model.add(Conv2D(64, 3, activation='relu'))  # 64 filters, 3x3 kernel size, ReLU activation
    
    model.add(MaxPooling2D(2, 2))  # Max pooling layer to downsample the feature maps
    
    
    
    # Flatten the feature maps to feed into the fully connected layers
    
    model.add(Flatten())  # Convert 2D matrix to 1D vector
    
    
    
    # Add fully connected layers
    
    model.add(Dense(512, activation='relu'))  # Fully connected layer with 512 units and ReLU activation
    
    model.add(Dense(64, activation='relu'))    # Fully connected layer with 64 units and ReLU activation
    
    model.add(Dense(5, activation='softmax'))   # Output layer for 5 classes with softmax activation
```

```python
# Display the summary of the model

model.summary()
```

```python
# Plot the defined model using keras.utils.plot_model

keras.utils.plot_model(

    model,                        # A Keras model instance to be visualized

    to_file="model_structure.png",         # The filename to save the plot (with format "name.format")

    show_shapes=True,            # Display shape information of each layer

    show_dtype=False,            # Show the data types of the layers (default is False)

    show_layer_names=True,       # Display the names of the layers in the plot

    rankdir="TB",                # Direction to layout the graph: 'TB' for top-to-bottom

    expand_nested=False,         # Expand nested Functional models into clusters (default is False)

    dpi=200,                     # Image resolution in dots per inch

    show_layer_activations=True, # Display the activation functions for each layer (default is False)

    show_trainable=False          # Indicate whether to show if a layer is trainable (default is False)

)
```

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 4.2 | </b></span><span style="color:#1d84b5"><b>Configure Callbacks</b></span>


<div style="background-color:#0A2342; padding: 25px 10px 10px 10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.1); border:0px solid #0A2342; text-align:left; color: #f2f2f2;">
    <p style="font-size:18px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        ➡️ A <b>callback</b> is an object that can perform actions at various stages of training 
        (e.g., at the start or end of an epoch, before or after a single batch, etc).
    </p>
    <p style="font-size:18px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        You can use callbacks to:
        <ul style="margin-left: 40px; font-size:18px; line-height: 1.8em;">
            <li>Write TensorBoard logs after each training batch to monitor your metrics.</li>
            <li>Periodically save your model to disk.</li>
            <li>Implement early stopping.</li>
            <li>Observe internal states and model statistics during training.</li>
            <li>...and more.</li>
        </ul>
    </p>

</div>


```python
# Model CheckPoint Call-Back: Saves the best model parameters as a .keras file

checkpoint_cb = keras.callbacks.ModelCheckpoint(

    'CNN_Image_Classification.keras',         # Name of the file to save the model

    save_best_only=True      # Only save the model if it's the best observed so far

) 



# ReduceLROnPlateau Call-Back: Reduces the learning rate when a monitored metric has stopped improving

reduce_lr = keras.callbacks.ReduceLROnPlateau(

    monitor='val_loss',      # Metric to monitor (validation loss in this case)

    factor=0.5,              # Factor by which to reduce the learning rate (new_lr = old_lr * factor)

    patience=3,              # Number of epochs with no improvement after which learning rate will be reduced

    min_lr=1e-6              # Lower bound on the learning rate

)
```

<a id='step43'></a>

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 4.3 | </b></span><span style="color:#1d84b5"><b>Compile Model</b></span>


```python
# 3: Build and compile the model within the strategy scope
# Using the strategy.scope() ensures that the model is distributed across multiple GPUs
with strategy.scope():
    # Compile the model with Adam optimizer
    model.compile(
        loss='categorical_crossentropy',              # Loss function for multi-class classification
        optimizer=optimizers.Adam(learning_rate=0.001),  # Adam optimizer with a learning rate of 0.001
        metrics=['accuracy']                          # Metric to monitor during training
    )
```

<a id='step44'></a>

## <span style="font-family:tahoma;font-size:100%;text-align:left"><span style="color:red"><b>Step 4.4 | </b></span><span style="color:#1d84b5"><b>Train Model</b></span>


<div style="background-color:#0A2342; padding: 25px 10px 10px 10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.1); border:0px solid #0A2342; text-align:left; color: #f2f2f2;">
    <p style="font-size:18px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        ➡️ Here, to train the model, we choose <b>epochs = 25 </b>. By defining appropriate callbacks, this may not require the model to train for all epochs.
    </p>

</div>


```python
num_epochs = 10  # Define the number of epochs (iterations over the entire dataset) for training

# Train the model using the training and validation data
history = model.fit(
    train_generator,              # The data generator for training data, which yields batches of images and labels
    epochs=num_epochs,            # Specify the number of epochs for training
    validation_data=test_generator,  # The data generator for validation data, used to monitor performance during training
    callbacks=[checkpoint_cb, reduce_lr]  # Callbacks for specific actions during training:
                                          # 1. checkpoint_cb: saves the best model based on validation loss
                                          # 2. reduce_lr: adjusts learning rate when training reaches a plateau
)
```

<div style="background-color:#0A2342; padding: 25px 10px 10px 10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.1); border:0px solid #0A2342; text-align:left; color: #f2f2f2;">
    <p style="font-size:18px; font-family:tahoma; line-height: 2em; text-indent: 20px;">
        ➡️ After training the model, all saved losses, accuracies, and learning rates are stored in a variable. This can be converted to a DataFrame for easy plotting.
    </p>

</div>


```python
# Convert the results of training into a DataFrame for easier analysis

result_df = pd.DataFrame(history.history)



# Display the DataFrame containing the training history

print(result_df)
```

```python
# Define a variable x to store the range of epochs for plotting
x = np.arange(len(result_df))

# Create a figure with three subplots for accuracy, loss, and learning rate
fig, ax = plt.subplots(3, 1, figsize=(10, 18))

# ax[0] ---> Accuracy
ax[0].plot(x, result_df.accuracy, label='Train Accuracy', linewidth=2)  # Plot training accuracy
ax[0].plot(x, result_df.val_accuracy, label='Validation Accuracy', linewidth=2, ls='-.', c='r')  # Plot validation accuracy
ax[0].set_title('Accuracy')  # Set the title for the accuracy plot
ax[0].set_yticks(np.arange(0.95, 1.01, 0.01))  # Set y-ticks for accuracy
ax[0].set_xticks(np.arange(0, num_epochs, 1))  # Set x-ticks for epochs
ax[0].legend()  # Show legend for accuracy plot
ax[0].tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees

# ax[1] ---> Loss
ax[1].plot(x, result_df.loss, label='Train Loss', linewidth=2)  # Plot training loss
ax[1].plot(x, result_df.val_loss, label='Validation Loss', linewidth=2, ls='-.', c='r')  # Plot validation loss
ax[1].set_title('Loss')  # Set the title for the loss plot
ax[1].set_xticks(np.arange(0, num_epochs, 1))  # Set x-ticks for epochs
ax[1].set_yticks(np.arange(0, 0.085, 0.01))  # Set y-ticks for loss
ax[1].legend()  # Show legend for loss plot
ax[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees

# ax[2] ---> Learning Rate
ax[2].plot(x, result_df.learning_rate, label='Learning Rate', linewidth=2, marker='o')  # Plot learning rate
ax[2].set_title('Learning Rates')  # Set the title for the learning rate plot
ax[2].set_xticks(np.arange(0, num_epochs, 1))  # Set x-ticks for epochs
ax[2].legend()  # Show legend for learning rate plot
ax[2].tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees

# Display the plots
plt.show()
```

```python
# Reset the test generator to ensure we start from the beginning of the dataset

test_generator.reset()



# Get the next batch of images and labels from the test generator

img, label = next(test_generator)



# Predict the classes for the current batch of images

prediction = model.predict(img)



# Get the predicted class indices by taking the argmax across the predicted probabilities

test_pred_classes = np.argmax(prediction, axis=1)



# Set up the figure for displaying the images and their predictions

plt.figure(figsize=[14, 14])

for i in range(16):  # Loop through the first 16 images in the batch

    plt.subplot(4, 4, i + 1)  # Create a subplot for each image

    plt.imshow(img[i])  # Display the image

    plt.axis('off')  # Hide the axis

    plt.title("Label : {}\n Prediction : {} {:.1f}%".format(

        classes[np.argmax(label[i])],  # Get the true label

        classes[test_pred_classes[i]],  # Get the predicted label

        100 * np.max(prediction[i])  # Get the confidence of the prediction

    ))

plt.show()  # Display all the images and their predictions
```

```python
# Predict the classes for all images in the test generator

y_pred = model.predict(test_generator, verbose=1)



# Convert predicted probabilities to class labels

y_pred_bool = np.argmax(y_pred, axis=1)



# Generate a classification report comparing true classes to predicted classes

clf_report = classification_report(test_generator.classes, y_pred_bool, target_names=classes)



# Print the classification report

print(clf_report)
```
